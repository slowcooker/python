SIN波を描いてみた。

```python
import numpy as np
from matplotlib import pylab as plt
import math # 抜けてた

sinWave = []

smoothness = 5 # 5度ずつ傾ける。

for n in range(0,100):
    #np.sin(rad) rad は角度
    deg = smoothness * n 
    # 度をラジアンに変換
    rad = math.radians(deg)
    s = np.sin(rad)
    sinWave.append(s)
 
# サイン波を表示
print("サイン波。")
plt.plot(sinWave[0:100])
plt.show()

cos = []

for n in range(0,100):
    #np.sin(rad) rad は角度
    deg = smoothness * n 
    # 度をラジアンに変換
    rad = math.radians(deg)
    s = np.cos(rad)
    cos.append(s)
 
# コサイン波を表示
print("コサイン波。")
plt.plot(cos[0:100])
plt.show()

tan = []

for n in range(0,100):
    #np.sin(rad) rad は角度
    deg = smoothness * n 
    # 度をラジアンに変換
    rad = math.radians(deg)
    s = np.tan(rad)
    tan.append(s)
 
# タンジェント波を表示
print("タンジェント波。…これでいいんだっけ？")
plt.plot(tan[0:100])
plt.show()


```

１つの表にまとめる。

```python
import matplotlib.pyplot as pyp
fig = pyp.figure()
ax = fig.add_subplot(1,1,1)
x = np.linspace(0, 100)
ax.plot(sinWave[0:100], color='red',  linestyle='solid')
ax.plot(cos[0:100], color='blue',  linestyle='solid')
ax.set_title('Combined Waves')
ax.set_xlabel('args x position')
ax.set_ylabel('amplitude')
ax.grid(True)
```

```python
sin2 = []
for n in range(0,100):
    #np.sin(rad) rad は角度
    deg = smoothness * n 
    # 度をラジアンに変換
    rad = math.radians(deg)
    s = np.sin(rad)
    # sにY軸に対して10%以内のノイズを加える。 
    s += np.random.uniform(-1.0, 1.0) * 0.1
    sin2.append(s)

fig = pyp.figure()
ax = fig.add_subplot(1,1,1)
x = np.linspace(0, 100)
ax.plot(sinWave[0:100], color='red',  linestyle='solid')
ax.plot(sin2[0:100], color='blue',  linestyle='solid')
ax.set_title('Combined Waves')
ax.set_xlabel('args x position')
ax.set_ylabel('amplitude')
ax.grid(True)
```
